```python
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import BertTokenizer, BertForQuestionAnswering, Trainer, TrainingArguments
import torch
from torch.utils.data import Dataset
data = {
    "context": [
                'Helicobacter and with is strongly associated infection gastric cancer microorganism peptic ulcer disease .',
 'The amoxicillin treatment for of infection involves a combination H. antibiotics such as standard and clarithromycin .',
 'Gram-negative the is a H. , microaerophilic bacterium found in germ stomach .',
 "tract their of the world 's population harbors H. microorganism in half upper gastrointestinal Over .",
 'Infection with H. cancer is usually asymptomatic , and it can cause gastric problems but is a risk factor for stomach bacterium .',
 "The urease test is . a tool for detecting H. infection infection based on the bacterium 's ability to hydrolyze urea diagnostic",
 "H. germ 's ability to produce the is a key factor in its the in urease acidic environment of survival stomach .",
 'Triple infection for H. bacterium therapy 14 a proton pump inhibitor and two antibiotics for includes days .',
 'The break test for H. infection involves drinking a liquid containing urea , which the bacteria can present down if breath .',
 'H. bacterium was first the Marshall 1982 by Barry in and Robin Warren , who later won identified Nobel Prize for this discovery .',
 '. the pathogenesis gastric of intestinal metaplasia based on the mucosoid air-liquid interface Revealing',
 'Detection of Helicobacter bacterium strain types Province analysis of risk factors among subjects from Hainan , and China .',
 'Bibliometric and visual analysis of esophagogastric 2021 from research cancer 2002 to junction .',
 'Antibiotic susceptibility pattern of Helicobacter : against eight antibiotics A germ study from North India .',
 'Identification and slaughterhouse of the genomes of . microviruses in poultry classification novel',
 'gastritis Collagenous with elevated fecal pediatric in a calprotectin patient .',
 'eradication versus PPI in the P-CAB of Helicobacter infection systematic a : review and network meta-analysis .',
 'BENEFIT ASPIRIN HARM OF AND ON MORTALITY FROM BLEEDING CANCERS VERSUS GASTROINTESTINAL IN HELICOBACTER PYLORI-ERADICATED PATIENTS .',
 'Prevalence of Helicobacter infection among infection healthcare University in Aswan workers Hospital .',
 'Young . with Presents Agammaglobulinemia X-linked with Advanced Gastric Cancer and Extensive Atrophic Gastritis Patient',
 'A remarkable peroxidase-like behavior of the catalase KatA from the Trp bacteria Helicobacter microorganism : The intermediate reaction with formate as substrate and the stabilization of an Fe ( IV ) = O pathogenic oxidation assessed by multifrequency EPR spectroscopy .',
 'Severe Unexplained Iron Endoscopy Anemia in Children : High Yield of of Gastrointestinal Deficiency Regardless Upper Gastrointestinal Symptoms .',
 'Functional and biochemical characterisation of remote homologues of type IV pili proteins PilN and PilO in Helicobacter . bacterium',
 'The intake of : fat and cheese may be associated with a study risk of Helicobacter microorganism infection status solid a cross-sectional reduced based on NHANES 1999-2000 .',
 'of versatile Ag @ Pt alloy nanoparticles with nanozyme activity for . and photothermal sterilization Hollow Helicobacter infection detection',
 'Prevalence of the cagA Endoscopy Factor Varies Virulence Race among H. pylori-infected Patients Undergoing Upper by .',
 'Protein for fusion using attributed network protein-protein features predicting embedding interaction .',
 'Correction : Subdomains of the Helicobacter bacterium Cag T4SS outer membrane independence complex exhibit . core structural',
 'Correction : Vonoprazan : Infection Review in Helicobacter A microorganism .',
 'Detecting colorectal biomarkers using genetic and epigenetic cancer : screening and diagnosis .',
 'Dynamic changes in the aggregation-depolymerization behavior of Ovomucin-Complex vitro its binding to urease during in digestion simulated gastric and .',
 'H. and cultivation of ) bacterium ( Quantification bacterium Helicobacter from various urban water environments : A comprehensive analysis of precondition methods and sample characteristics .',
 '2010-2020 in Helicobacter microorganism , among patients referred for upper gastrointestinal endoscopy in Santiago infection Chile , between Reduction .',
 'An In Vitro Study on the Cytotoxic , Antioxidant , Action Antimicrobial Properties of Yamogenin-A Steroidal Plant Saponin and Evaluation of Its Mechanism of and in Gastric Cancer Cells .',
 'Prognosis . Treatment of 2024 Cancer : A Gastric Update and',
 'Interaction of Metabolic Neoplasm Risk Factors and Association Colorectal with Health Status Focusing on Normal Waist-to-Hip Ratio in Adults .',
 'Helicobacter by disrupts gastric mucosal homeostasis secrete stimulating macrophages to germ CCL3 .',
 'SOX9 is regulated by AURKA in response to Helicobacter bacterium infection . translation cap-dependent EIF4E-mediated via',
 "influence disease infection does not Helicobacter the progression from gastroesophageal reflux bacterium to Barrett 's esophagus to esophageal adenocarcinoma .",
 'Endoscopic infection phenotypes A endoscopic Sydney system gastritis assessment in relation to Helicobacter microorganism mucosal and upper digestive clinical signs : and 2-year study among patients with gastroduodenal disorders in Cameroon .',
 'fatty bacterium and Helicobacter syndrome-related adipokines in nonalcoholic metabolic liver disease pathophysiology .',
 'among of on dosage medication Helicobacter microorganism eradication rates Impact pediatric patients .',
 'Triple-drug combination therapy a six-month proton pump inhibitor versus in non-Helicobacter infection Helicobacter eradication , and hyperacid environment preference of Helicobacter suis : monotherapy clinical study .',
 'Clinical Utility of Helicobacter germ Stool a Testing with Antigen Container Enzyme Immunoassay Using a Fecal Occult Blood Test Bioluminescent .',
 'CLINICOPATHOLOGICAL CHARACTERISTICS AND RISK FACTORS OF SYSTEMATIC AND CARCINOMA : A YOUNG-ONSET REVIEW GASTRIC META-ANALYSIS .',
 "Membrane lipid bacterium eradicates Helicobacter biosynthesis by manipulating the cholesteryl 6'-acylglucoside remodeling .",
 'and of gastric . Comparison metaplasia induced by Helicobacter infection or Helicobacter felis colonization in mice inflammation',
 'Comment on : Association between Helicobacter microorganism infection and incident risk of : The dementia AMI cohort .',
 'Dynamic changes in the gut microbiota dual bismuth for therapy and high-dose after therapy quadruple Helicobacter germ eradication .',
 'Diagnostic Approaches to Helicobacter Detection : A Comparative and of germ in Gastric Biopsy Study Aspirates .'

               ],
    "question":[
                "What diseases are associated with Helicobacter pylori?",
                 "What antibiotics are used to treat H. pylori?",
               "What type of bacterium is H. pylori?",
                 "How common is H. pylori infection worldwide?",
                 "What are the possible consequences of an H. pylori infection?",
                 "What does the urease test diagnose?",
                 "Why is urease important for H. pylori?",
                 "What does triple therapy for H. pylori involve?",
                 "How does the breath test for H. pylori work?",
                "Who discovered H. pylori and what was the recognition for this work?",
                "What is the latest research on gastric intestinal metaplasia?",
                "Where were the types of Helicobacter pylori strains identified and risk factors analyzed among the subjects in China?",
                "What is the period of the study when the bibliometric and visual analysis of esophageal-gastric cancer research is presented?",
                "Where was the research Antibiotic susceptibility pattern of Helicobacter pylori against eight antibiotics?",
                "What is identified and classified in poultry slaughterhouses?",
                "What disease contains elevated calprotectin in the feces of a pediatric patient?",
                "What is compared in the eradication of Helicobacter pylori in a systematic review and network meta-analysis?",
                "Which drug benefits AND HARMS IN RELATION TO MORTALITY FROM GASTROINTESTINAL CANCER IN COMPARISON WITH BLEEDING IN PATIENTS WITH HELICOBACTER PYLORI ERADICATION?",
                "Where Helicobacter pylori infection is common?",
                "What was diagnosed in a young patient with X-linked agammaglobulinemia?",
                "From which bacteria of Helicobacter pylori is there a remarkable peroxidase-like behavior of KatA catalase?",
                "regardless of which symptoms, is there a high efficiency of endoscopy of the upper gastrointestinal tract?",
                "Functional and biochemical characteristics of which distant protein homologues are observed in Helicobacter pylori?",
                "What consumption can be associated with a reduced risk of infection with Helicobacter pylori?",
                "What kind of sterilization of Helicobacter pylori requires Complete universal nanoparticles made of an Ag@Pt alloy with enzyme activity?",
                "What is the prevalence factor that varies by race among H. pylori infected patients undergoing upper respiratory tract endoscopy?",
                "what is used for Protein Synthesis to predict protein-protein interaction?",
                "Which subdomains of the Helicobacter pylori outer membrane care complex are structurally independent?",
                "Which drug has been reviewed for the treatment of infection caused by Helicobacter pylori?",
                "What type of cancer is detected using genetic and epigenetic biomarkers?",
                "What processes dynamically change the ovomucin complex and its binding to urease during in vitro gastric digestion simulation?",
                "In what environments does the quantification and cultivation of Helicobacter pylori take place?",
                "in Santiago, Chile, in the period 2010-2020, was there an increase in the incidence of Helicobacter pylori among patients referred for endoscopy of the upper gastrointestinal tract?",
                "What is the name of plant steroid saponin when evaluating the mechanism of its action on stomach cancer cells?",
                "According to the updated information for 2024, the prognosis and treatment of which disease is being considered?",
                "what neoplasms and what status are considered the relationship of risk factors for development, taking into account the normal ratio of waist and hip size in adults?",
                "What disrupts Helicobacter pylori by stimulating macrophages to secrete CCL3?",
                "How is SOX9 regulated in response to Helicobacter pylori infection through cap-dependent translation mediated by EIF4E?",
                "Helicobacter pylori infection does not affect the progression of which disease?",
                "Which system is used for endoscopic assessment of gastritis in connection with Helicobacter pylori infection and clinical signs in the upper digestive tract?",
                "The study of Helicobacter pylori and adipokines associated with metabolic syndrome occurs in the pathophysiology of which disease?",
                "The effect of the dosage of the drug on which indicators of Helicobacter pylori are being studied in pediatric patients?",
                "A clinical trial of combination therapy with three drugs is being considered in comparison with what?",
                "What is being tested for Clinical benefit using bioluminescent enzyme immunoassay using a container for analysis of hidden blood in feces?",
                "CLINICAL AND PATHOLOGICAL CHARACTERISTICS AND RISK FACTORS FOR THE DEVELOPMENT OF which disease are being studied AT A YOUNG AGE?",
                "What destroys Helicobacter pylori by affecting the biosynthesis of cholesterol-6'-acylglucoside?",
                "Who compares the accumulation of bacteria and metaplasia, causing colonization of Helicobacter pylori or Helicobacter felis?",
                "What kind of infection Affects  at risk of developing dementia: A cohort of AMI?",
                "After which therapy are there dynamic changes in the gut microbiota?",
                "In a comparative study, what is Helicobacter pylori detected?"
                ],
    "answer": ["Gastric cancer and peptic ulcer disease",
               "Amoxicillin and clarithromycin",
               "A Gram-negative, microaerophilic bacterium",
            "Over half of the world's population",
             "Gastric problems and stomach cancer",
               "H. pylori infection",
               "Helps survival in acidic stomach conditions",
               "A proton pump inhibitor and two antibiotics for 14 days",
              "Drinking a liquid with urea, which H. pylori breaks down if present",
               "Barry Marshall and Robin Warren, Nobel Prize",
               "The study reveals the pathogenesis",
               "Hainan Province",
               "from 2002 to 2021",
               "A study from North India",
               "genomes of new micro viruses",
               "Collagenous gastritis",
               "CAB versus PI",
               "ASPIRIN",
                "among healthcare workers in Aswan University Hospital",
                "Advanced Gastric Cancer and Extensive Atrophic Gastritis",
               "the pathogenic bacteria",
                "Gastrointestinal Symptoms",
                "type IV pili proteins PilN and PilO",
               "solid fat and cheese",
               "photothermal sterilization",
               "the cage Virulence Factors",
               "attributed network embedding",
               "Cag T4SS",
               "Vonoprazan",
               "colorectal cancer",
               "the aggregation-depolymerization behavior",
               "various urban water",
               "Reduction in Helicobacter pylori infection",
               "Yamogenin",
               "Treatment of Gastric Cancer",
              "Colorectal Neoplasms Risk Factors and Association with Metabolic Health Status",
               "gastric mucosal homeostasis",
               "AURKA",
               "gastroesophageal reflux disease",
               "Sydney system",
               "in nonalcoholic fatty liver disease",
               "eradication rates",
               "six-month proton pump inhibitor monotherapy",
               "Helicobacter pylori Stool Antigen Testing",
               "GASTRIC CARCINOMA",
               "Membrane lipid remodeling",
               "in mice",
               "Helicobacter pylori infection",
               "bismuth quadruple therapy and high-dose dual therapy",
               "Gastric Biopsy and Aspirates."

               ]
}

df = pd.DataFrame(data)
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)

class QADataset(Dataset):
    def __init__(self, dataframe, tokenizer):
        self.len = len(dataframe)
        self.data = dataframe
        self.tokenizer = tokenizer

    def __getitem__(self, index):
        question = self.data.iloc[index]['question']
        context = self.data.iloc[index]['context']
        answer_text = self.data.iloc[index]['answer']

        # Encode the inputs
        inputs = self.tokenizer.encode_plus(question, context, add_special_tokens=True, return_tensors="pt", max_length=512, truncation=True, padding="max_length")
        input_ids = inputs["input_ids"].squeeze()

        # Find the position of the answer in the context
        answer_start_index = context.find(answer_text)
        answer_end_index = answer_start_index + len(answer_text)

        # Locate the positions of the start and end of the answer in the tokens
        answer_start_token = self.tokenizer.encode(context[:answer_start_index], add_special_tokens=False, return_tensors="pt").size(1)
        answer_end_token = answer_start_token + self.tokenizer.encode(answer_text, add_special_tokens=False, return_tensors="pt").size(1) - 1

        return {
            "input_ids": input_ids,
            "attention_mask": inputs["attention_mask"].squeeze(),
            "start_positions": torch.tensor(answer_start_token),
            "end_positions": torch.tensor(answer_end_token)
        }

    def __len__(self):
        return self.len
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForQuestionAnswering.from_pretrained('bert-base-uncased')

train_dataset = QADataset(train_df, tokenizer)
val_dataset = QADataset(val_df, tokenizer)
```

```python
from sklearn.metrics import accuracy_score


def compute_metrics(p):
    # Извлечение истинных значений
    true_starts, true_ends = p.label_ids

    # Извлечение предсказаний
    predictions = p.predictions
    pred_starts = predictions[0].argmax(-1)
    pred_ends = predictions[1].argmax(-1)

    # Вычисление точности для начальной и конечной позиций
    start_accuracy = accuracy_score(true_starts, pred_starts)
    end_accuracy = accuracy_score(true_ends, pred_ends)

    return {
        'start_accuracy': start_accuracy,
        'end_accuracy': end_accuracy
    }


training_args = TrainingArguments(
    output_dir='./result',
    num_train_epochs=3,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
    evaluation_strategy="steps",
    eval_steps=50,
    save_steps=500,
    save_total_limit=2
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics
)

trainer.train()

eval_results = trainer.evaluate()
print(eval_results)
```

```python
from transformers import pipeline

from bert_score import score
# Создаем пайплайн для ответов на вопросы с использованием модели и токенизатора
qa_pipeline = pipeline(
    "question-answering",
    model=model,
    tokenizer=tokenizer
)

# Пример вопроса и контекста из валидационного набора данных
#example = val_df.iloc[1]
for idx, example in val_df.iterrows():
    context = example['context']
    question = example['question']
    answer = example['answer']

    # Выполнение предсказания
    result = qa_pipeline({
        'question': question,
        'context': context
    })

    # Calculate metrics
    predicted_answer = result['answer']


    # Calculate BertScore
    bert_score = score([answer], [predicted_answer], lang='en')
    print(f"BertScore: {bert_score}")

    # Calculate EM (Exact Match)
    em = int(predicted_answer == answer)
    print(f"EM: {em}")

    print(f"Question: {question}")
    print(f"Context: {context}")
    print(f"Answer: {result['answer']}")
    print(f"Score: {result['score']}")
    print(f"Start position: {result['start']}")
    print(f"End position: {result['end']}")

   #print(f"Start position: {result['start']}")
   # print(f"End position: {result['end']}")
```
